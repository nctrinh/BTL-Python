```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

# <font size="6">1. Khai báo tên file</font>

```python
languages_file_name = r'D:\projects\Python\BTL\raw_data\languages_data.json'
```

# <font size="6">2. Mở dữ liệu và làm sạch</font>

```python
with open(languages_file_name, 'r', encoding='utf-8') as file:
    languages_data = json.load(file)
languages_data = languages_data[0]
languages_series = pd.Series(languages_data.values(), index=languages_data.keys(), name='Size(Byte)')
```

```python
languages_series.to_excel(r'D:\projects\Python\BTL\cleaned_data\cleaned_languages_data.xlsx')
```
