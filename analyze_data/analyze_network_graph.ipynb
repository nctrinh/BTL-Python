```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
commits_in_file_file_name = r'D:\projects\Python\BTL\cleaned_data\cleaned_commits_in_file_data.xlsx'
```

```python
df = pd.read_excel(commits_in_file_file_name, index_col=0)
df
```

```python
file_name_list = list(set(df['filename'].to_list()))
file_name_list = file_name_list[:200]
```

```python
contributor_name_list = list(set(df['commit.author.name'].to_list()))
len(contributor_name_list)
```

```python
files_contributions = {}
for file in file_name_list:
    file_mask = df['filename'] == file
    contributor_name_commit_in_file = list(set(df[file_mask]['commit.author.name'].to_list()))
    files_contributions[file] = contributor_name_commit_in_file
```

```python
G = nx.Graph()
for x in contributor_name_list:
    G.add_node(x)
```

```python
plt.figure(figsize=(50, 50))
pos = nx.random_layout(G)
nx.draw(G, with_labels=True, node_size=300, node_color='blue', font_size=10, font_weight='bold', edge_color='gray')
# plt.title(f'Network Graph of Contributors in {repo_owner}/{repo_name}', size=15)
plt.show()
```
